```python
import pandas as pd
# Load the cleaned CSV file
df = pd.read_csv('../project/assets/data/cleaned.csv')
# Display basic information
print(f"Dataset shape: {df.shape}")
df.head(2)
```

```python
# Import the model
from model import SimpleNBSRecommender, choose_columns_and_k_interactively

# -------------------- Example usage --------------------
# For testing without interaction, we'll use predefined columns and k
selected_cols = [
    "Country",
    "Present stage of the intervention",
    "Duration (End - Begin)",
    "NBS area (m2)",
    "Total cost €",
    "Type of area before implementation of the NBS",
    "Source(s) of funding",
]
k = 5

# Fit the recommender
rs = SimpleNBSRecommender().fit(df, selected_cols)

# Define user preferences
user_preferences = {
    "Country": "Turkey",
    "Present stage of the intervention": "Completed",
    "Duration (End - Begin)": 6,
    "NBS area (m2)": 200000,
    "Total cost €": 1500000,
    "Type of area before implementation of the NBS": "Public Greenspace",
    "Source(s) of funding": ["Public local authority budget", "Multilateral fund"],
}

# Get recommendations
results = rs.recommend(user_preferences, n_results=k)
print(f"\nTop {k} recommendations based on your preferences:")
print(results)
```
